# Amostragem Sistematica

```python
def amostragem_sistematica(dataset, amostras):
  intervalo = len(dataset) // amostras
  random.seed(5)
  inicio = random.randint(0, intervalo)
  indices = np.arange(inicio, len(dataset), step=intervalo)
  dataset = dataset.iloc[indices]
  return dataset
```

```python
dados_sistematica = amostragem_sistematica(dados, 1000)
dados_sistematica
```

```python
#ESSE
def amostragem_grupo(dados, grupos):
  random.seed(1)
  grupo_selecionado = random.randint(0,1)
  tamanho = len(dados) // grupos
  grupos = []
  id_grupos = 0
  contagem = 0
  for _ in dados.iterrows():
    grupos.append(id_grupos)
    contagem += 1
    if contagem > tamanho:
      contagem = 0
      id_grupos +=1
  dados['grupo'] = grupos
  dados_grupos = dados[dados['grupo'] == grupo_selecionado]
  return dados_grupos
```

```python
dados_grupos = amostragem_grupo(dados, 2)
dados_grupos
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
def amostragemestratificada(dataset, amostra, campo):
  porc = (amostra*100)/200000
  estratificados = StratifiedShuffleSplit(test_size=porc)
  for x, y in estratificados.split(dataset, dataset[campo]):
    df_x = dados.iloc[x]
    df_y = dados.iloc[y]
  return df_y
```

# AMostragem Reservatorio

```python
def Amostragem_reservatorio(dataset, amostras):
  stream = []
  for i in range(len(dataset)):
    stream.append(i)

  i = 0
  tamanho = len(dataset)

  reservatorio = [0] * amostras
  for i in range(amostras):
    reservatorio[i] = stream[i]

  while i < tamanho:
    j = random.randrange(i + 1)
    if j < amostras:
      reservatorio[j] = stream[i]
    i += 1

  return dataset.iloc[reservatorio]

```

```python
dados_reservatorio = Amostragem_reservatorio(dados, 1000)
dados_reservatorio
```
