# Dados absolutos e relativos

```python
import pandas as pd
import numpy as np
from array import array
```

### Percentuais

```python
dados = {'emprego': ['Adminstrador_banco_dados','Programado', 'Arquiteto_redes'],
         'nova_jersey' : [97350, 82080, 112840],
         'florida' : [77140, 71540, 62310]}
```

```python
type(dados)
```

```python
dados
```

```python
dataset = pd.DataFrame(dados)
dataset
```

```python
print(f"Total em Nova Jersey: {dataset['nova_jersey'].sum()}")
print(f"Total na Florida: {dataset['florida'].sum()}")
```

```python
dataset['%_nova_jersey'] = (dataset['nova_jersey'] / dataset['nova_jersey'].sum() * 100)
dataset
```

```python
dataset['%_florida'] = (dataset['florida'] / dataset['florida'].sum() * 100)
dataset
```
